```python
#import os
#os.chdir("/home/gianniif/ecommerce-predictor/")
from utils.build_df import build_df
import requests
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python

df = build_df(json_path='products.json', 
             threshold=100, 
             preprocessed_csv=None
            )

#adding the ending column
ending= []

for row in df['image']:
    ending.append(row.split('.')[-1])

df['ending'] = ending

#deleting gift and broken urls
ind = 0  
for row in df['image']:        
    image = requests.get(row).content
    if len(image) == 9 or df['ending'][ind] == 'gif':
        df2 = df.drop(index =ind, axis=0 ,inplace=True)
        print('delete',ind)
    ind +=1

#Splitting train/test
y=[]
X = []

for index ,row in df.iterrows():
    X.append("row" + str(index+1) +'.'+ str(df.ending[index]))
    y.append(df.category[index][-1]['id'])


X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20, 
    random_state=42,
    stratify = y
)

len(X_train)

#creating the columns for the cvs
img_name = X_train + X_test
subset = ['train']*len(X_train) + ['test']*len(X_test)
cat = y_train + y_test

#creating the cvs
data= [img_name,cat,subset]
df = pd.DataFrame(data).transpose()
df = df.rename(columns={0: 'img_name', 1: 'class', 2: 'subset'})

df.to_csv('prod_dataset_labels.csv', index=False)

print (df.groupby('subset')['class'].nunique())
```
